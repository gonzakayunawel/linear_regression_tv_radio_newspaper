```python
# Gráfico de dispersión de TV vs Ventas
x_tv = data['tv']
target = data['ventas']

plt.figure(figsize=(10, 6))
plt.scatter(x_tv, target, color='tomato')
plt.show()
```

```python
# Separar datos de entramiento y datos de prueba
from sklearn.model_selection import train_test_split

X_tv_train, X_tv_test, y_tv_train, y_tv_test = train_test_split(x_tv, target, test_size=0.2)

X_tv_train = X_tv_train.to_numpy().reshape((-1,1))
X_tv_test = X_tv_test.to_numpy().reshape((-1,1))
y_tv_train = y_tv_train.to_numpy().reshape((-1,1))
y_tv_test = y_tv_test.to_numpy().reshape((-1,1))
```

```python
# Desde la librería sklearn.linear_model importamos LinearRegression
from sklearn.linear_model import LinearRegression

plt.figure(figsize=(14, 8))

# Creamos una variable llamada model con la Regresión Lineal

model_tv = LinearRegression()

# Le indicamos los parámetros para entrenamiento
model_tv.fit(X_tv_train,y_tv_train)

# Le pedimos que calcule el índice de correlación R de Pearson
R_sq_tv = model_tv.score(X_tv_train,y_tv_train)

print("El coeficiente de determinación R de Pearson es: ", round(R_sq_tv, 2))
print("El intercepto Beta_0 es: ", model_tv.intercept_)
print("La pendiente Beta_1 es: ", model_tv.coef_)
print("La función es estimada es: {0} + {1}x ".format(model_tv.intercept_, model_tv.coef_))

# Haremos una predicción
target_pred = model_tv.predict(X_tv_test)

plt.scatter(X_tv_train, y_tv_train, color='tomato', label='Datos de entrenamiento')
plt.scatter(X_tv_test, y_tv_test, color='b', label='Datos de prueba')
plt.plot(X_tv_test, target_pred, marker='.', linestyle='-', color='g', label='Datos predichos')

plt.title('Modelo de Regresión Datos de TV', fontsize=15)
plt.legend(fontsize=12)
plt.xlabel('Presupuesto TV en M', fontsize=12)
plt.ylabel('Ventas en M', fontsize=12)
plt.show()
```
